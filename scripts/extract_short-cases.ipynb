```python
import fitz
from operator import itemgetter
import re
filename = "../data/medicine_books/short_cases_medicine.pdf"
doc = fitz.open(filename)
```

```python
start = 177
end = 203
```

```python
ques = [""]
ans = []

q = False
a = False

for page in doc.pages(start,end):
    blocks = page.get_text("dict", sort = True)["blocks"]
    # questions = []
    # q = False
    # a = False
    for b in blocks:  # iterate through the text blocks
        if b['type'] == 0:  # block contains text
            for l in b["lines"]:  # iterate through the text lines
                for s in l["spans"]:  # iterate through the text spans

                    # print(s)
                    bold = "Bold" in s['font']
                    text = s['text'].strip()
                    hasq = "QUESTION" in text
                    # print(text)

                    if text:
                        if not q:
                            if not a:
                                if bold:
                                    if hasq:
                                        a = False
                                        q = True
                                    else:
                                        q = False
                                        a = False
                            else:
                                if bold:
                                    a = False
                                    if not re.search(r"[A-Z]+[A-Z]+",text):
                                        q = True
                                        ques.append(text)
                                    elif hasq:
                                        q = True
                                    else:
                                        q = False
                                else:
                                    ans[-1] += text
                        else:
                            if bold:
                                ques[-1] += text
                            else:
                                q = False
                                a = True
                                ques.append("")
                                ans.append(text)
```

#### Preparing History attached QA data

```python
with open("../extracted_text/short_cases_medicine/questions.txt", "r") as f:
    ques = f.readlines()

with open("../extracted_text/short_cases_medicine/answers.txt", "r") as f:
    ans = f.readlines()

with open("../extracted_text/short_cases_medicine/medical_context.txt", "r") as f:
    hist = f.readlines()
```

```python
def strip_newline(val):
    return val.strip("\n")
```

```python
hist = " ".join(list(map(strip_newline, hist))).split("||")
```

```python
len(hist)
```

```python
resp_indices = [25,4,14,20,15,4,11,9,22,13,3,9,7,3,3]
endo_indices = [2,19,9,18,8,9,9,6,4,6,3,9,6]
```

```python
print(len(resp_indices))
print(len(endo_indices))
```

```python
data = {}
resp = []
endo = []
ptr = 0

for i in range(len(resp_indices)):
    hs = hist[i]
    for j in range(resp_indices[i]):
        q = ques[ptr + j]
        a = ans[ptr + j]
        resp.append({"med_hist" : hs, "ques" : q, "ans" : a})
    ptr += resp_indices[i]

for i in range(len(endo_indices)):
    hs = hist[i + len(resp_indices)]
    for j in range(endo_indices[i]):
        q = ques[ptr + j]
        a = ans[ptr + j]
        endo.append({"med_hist" : hs, "ques" : q, "ans" : a})
    ptr += endo_indices[i]
```

```python
endo
```

```python
data["respiratory"] = resp
data["endocrine"] = endo
```

```python
import json
with open("../extracted_text/short_cases_medicine/short_cases_medicine_annotated.json", "w") as f:
    json.dump(data,f, indent = 4)
```
