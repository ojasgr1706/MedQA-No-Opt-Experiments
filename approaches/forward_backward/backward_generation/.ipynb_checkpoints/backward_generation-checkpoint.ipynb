```python
output_delimiter = "answer is ("

for i in range(len(gen_out)):
    out = gen_out.loc[i,'outputs']
    
    pred = ""
    pred_list = out.split(output_delimiter)
    if len(pred_list) > 1:
        pred_str = pred_list[1]
        if pred_str:
            pred = pred_str[0]
            
    gen_out.loc[i,"prediction"] = pred
```
